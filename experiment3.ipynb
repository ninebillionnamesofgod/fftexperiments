```python
import soundfile as sf
import scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft

%matplotlib inline

data, rate = sf.read("god.flac")

N = int(len(data)/200)
T = 1.0/1000.0

dfft = fft(data)

xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(dfft[:N//2]))
plt.show()
```

```python
from scipy.signal import spectrogram
specgram = spectrogram(data)

xs = np.linspace(0.0, 129.0, 129.0)
figa, axa = plt.subplots()
axa.plot(specgram[0], xs)
plt.show()
```
