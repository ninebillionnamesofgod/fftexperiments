```python
import soundfile as sf
import scipy
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft

%matplotlib inline

data, rate = sf.read("clip1.flac")
    
clip1fft = fft(data.T[0])

N = len(data)
T = 1.0 / 5000.0

xf = np.linspace(0.0, 1.0/(2.0*T), N/2)

fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(clip1fft[:N//2]))
plt.show()

norm = [(ele/2.**32)*2-1 for ele in data.T[0]]
normfft = fft(norm)

xxf = np.linspace(0.0, 1.0/(2.0*T), N/2)
figb, ay = plt.subplots()
ay.plot(data, np.abs(clip1fft))
plt.show()
```

```python
rfft = np.fft.rfftn(data.T[0])

figc, az = plt.subplots()
az.plot(data.T[0][:len(data)//2], np.abs(rfft[:len(data)//2]))
plt.show()

figd, aa = plt.subplots()
aa.plot(data.T[0][:N//2], np.abs(normfft[:N//2]))
plt.show()
```
