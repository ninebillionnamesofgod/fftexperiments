```python
import soundfile as sf
import librosa.feature
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline


plt.figure(figsize=(10,22))

def plot_mel(spec, pos, name):
    plt.subplot(12, 1, pos)
    librosa.display.specshow(librosa.power_to_db(spec, ref=np.max),
                             y_axis='mel', fmax=8000,
                             x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel spectrogram:{}'.format(name))
    plt.tight_layout()
    
i=1
for name in ["god", "god1", "god2", "god3", "god4", "god5", "god6", "god7", "noise", "yeah", "ohyeah", "oh-my-god"]:
    data, rate = sf.read("{}.flac".format(name))
    S = librosa.feature.melspectrogram(data, rate, n_mels=128, fmax=8000)
    plot_mel(S, i, name)
    i += 1
    
```

```python
plt.figure(figsize=(10,6))
def plot_cs(spec, pos, name):
    plt.subplot(2, 2, pos)
    librosa.display.specshow(spec, y_axis='chroma', x_axis='time')
    plt.colorbar()
    plt.title("Chromagram: {}".format(name))
    plt.tight_layout()
    
i=1
for name in ["god", "god1", "god2", "god3"]:
    data, rate = sf.read("{}.flac".format(name))
    S = librosa.feature.chroma_stft(data, rate)
    plot_cs(S, i, name)
    i += 1
```

```python
plt.figure(figsize=(10,6))

def plot_mel(spec, pos, name):
    plt.subplot(2, 2, pos)
    librosa.display.specshow(spec, x_axis='time')
    plt.colorbar()
    plt.title('MFCC: {}'.format(name))
    plt.tight_layout()
    
i=1
for name in ["god", "god1", "god2", "god3"]:
    data, rate = sf.read("{}.flac".format(name))
    S = librosa.feature.mfcc(data, rate, n_mfcc=40)
    plot_mel(S, i, name)
    i += 1
```

```python
plt.figure(figsize=(10,6))

def plot_spec_cent(spec, pos, name):
    plt.subplot(4, 2, pos)
    plt.semilogy(spec.T, label='Spectral centroid')
    plt.ylabel('Hz')
    plt.title("clip: {}".format(name))
    plt.xticks([])
    plt.xlim([0, spec.shape[-1]])
    plt.legend()
    
i=1
for name in ["god", "god1", "god2", "god3", "god4", "god5", "god6", "god7"]:
    data, rate = sf.read("{}.flac".format(name))
    S = librosa.feature.spectral_centroid(data, rate)
    plot_spec_cent(S, i, name)
    i += 1
    
    
```

```python
plt.figure(figsize=(10,6))

def plot_wave(spec, rate, pos, name):
    plt.subplot(4, 2, pos)
    librosa.display.waveplot(spec, rate)
    plt.ylabel("amp")
    plt.xlabel("time(s)")
    plt.title("waveform: {}".format(name))
    plt.legend()
    
i=1
for name in ["god", "god1", "god2", "god3", "god4", "god5", "god6", "god7"]:
    data, rate = sf.read("{}.flac".format(name))
    plot_wave(data, rate, i, name)
    i += 1
```
