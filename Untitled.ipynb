```python
import soundfile as sf
import numpy as np
import librosa.feature
import librosa.display
import librosa
import matplotlib.pyplot as plt

%matplotlib inline


def plot_mel(signal, rate, name):
    plt.figure(figsize=(10, 10))
    mels = librosa.feature.melspectrogram(signal, rate, n_mels=60)
    logspec = librosa.logamplitude(mels, ref_power=np.max)
    logspec = logspec.T.flatten()[:, np.newaxis].T
    #plt.semilogy(np.arange(0, 60, 1), logspec)
    librosa.display.waveplot(np.abs(mels), rate, x_axis='time')
    plt.title("Mel spectrum: (log) - {}".format(name))
    #plt.colorbar(format='%+02.0f dB')
    

data, rate = sf.read("god.flac")
plot_mel(data, rate, "god")

d1, r1 = sf.read("god1.flac")
plot_mel(d1, r1, "god1")

d2, r2 = sf.read("god2.flac")
plot_mel(d2, r2, "god2")
```
